```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leer el archivo de resultados
df = pd.read_csv("gname2.txt", delim_whitespace=True, header=None,
                 names=["tiempo", "Ep", "Ec", "E_total"])
df.head()
```

```python
# Graficar energ√≠a total
plt.plot(df["tiempo"], df["E_total"], label="Energ√≠a total")
#plt.plot(df["tiempo"], df["Ep"], label="Energ√≠a potencial")
#plt.plot(df["tiempo"], df["Ec"], label="Energ√≠a cin√©tica")

plt.xlabel("Tiempo")
plt.ylabel("Energ√≠a total")
plt.title("Evoluci√≥n de la energ√≠a en el tiempo")
plt.grid(True)
plt.minorticks_on()
#plt.legend()
plt.savefig("grafico4.png") 
plt.show()
```
