```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Constantes
k = 1  # Constante de Boltzmann (J/K)
m = 1      # Masa de una molécula de N2 (kg)
T = 1       # Temperatura (K)
```

```python
# Rango de velocidades
v = np.linspace(0, 10, 1000)
```

```python
# Distribución de Maxwell-Boltzmann
f_v = 4 * np.pi * (m / (2 * np.pi * k * T))**1.5 * v**2 * np.exp(-m * v**2 / (2 * k * T))
```

```python
# Gráfica
plt.figure(figsize=(8,5))
plt.plot(v, f_v, color='royalblue', label='Distribución de Maxwell-Boltzmann')
plt.title('Distribución de Maxwell-Boltzmann de velocidades')
plt.xlabel('Velocidad (m/s)')
plt.ylabel('Probabilidad (1/m/s)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Componente de la velocidad
vx = np.linspace(-5, 5, 1000)
```

```python
# Distribución de una componente vx
f_vx = np.sqrt(m / (2 * np.pi * k * T)) * np.exp(-m * vx**2 / (2 * k * T))
```

```python
# Graficar
plt.figure(figsize=(8,5))
plt.plot(vx, f_vx, color='darkred', label='Distribución de $v_x$')
plt.title('Distribución de Maxwell-Boltzmann para una componente de la velocidad')
plt.xlabel('$v_x$ (m/s)')
plt.ylabel('Probabilidad (1/m/s)')
plt.legend()
plt.grid(True)
plt.show()
```
